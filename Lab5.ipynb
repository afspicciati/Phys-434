Column Meanings:

pt - scalar transverse momentum. $\vec{p_T} \equiv (p_x,p_y)$, $p_T \equiv |p_T|$

eta - pseudorapidity. $\eta = ln cot{\frac{\theta}{2}}$ where $\theta$ is the polar angle.

phi - azimulathal angle of scattering. $\phi = \tan^{-1}{\frac{p_x}{p_y}}$

mass - invariant mass. Proportion of the mass that is independent of motion of the system. For a two particle collision this is given: $m = \sqrt{(E_1 + E_2)^2 - (\vec{p_1} + \vec{p_2})^2}$

ee2/ee3 - energies of the particles in the collision.

t_n - N-subjettiness. The degree to which jet radiation alligns along apecified subjet axes. Small values correspond to N or fewer subjets, large values correspond to N or more subjets.

t21 - ratio of t1 to t2, which "can identify boosted W/Z and Higgs bosons"

t32 - ratio of t3 to t2, which "is a successful boosted top discriminator"

KtDeltaR - $\Delta_R$ of two subjets within large-R jet

```python
def plot_both(data1,data2,attribute1,attribute2=None,bins=40):
    
    if attribute2 == None:
        attribute2 = attribute1
        
    fig, ax = plt.subplots(1, 2)
    plt.rcParams["figure.figsize"] = (25,7)
    plt.tick_params(labelsize = 10)
    
    ax[0].hist(data1[attribute1],bins=bins)
    ax[1].hist(data2[attribute2],bins=bins)

    
    ax[0].set_title(f'{attribute1} histogram for Background dataset')
    ax[1].set_title(f'{attribute2} histogram for Higgs dataset')

    ax[0].set_yscale('log')
    ax[1].set_yscale('log')
```

```python
for i in higgs.keys():
    plot_both(qcd,higgs,i)
```

# 1. Do all of the features provide discrimination power between signal and background?

Here I have plotted histograms for all features of the data for both the QCD simulated background data and the signal Higgs dataset. All are plotted in with a logarythmic y-axis to allow us to see detail in the data that would allow us to pick out difference in the signal in background. 

In general, the Higgs data is less evenly spread and more narrowly located than the QCD data. This makes sense, as the Higgs data has the actual signal, and not the entire range of possible signals. Now I will remark on the discrimination power between signal and background for all the features:

pt - The range is the same for this feature, but the Higgs data is more peaked, and the QCD data is more smooth. Higgs data has a peak at 1100 and falls off on either side. It is concave up on the left of the peak and concave down on the right of the peak. This is distinct from the QCD which has a flat peak around 1075 and falls off much more smoothly on either side.

eta - Near identical, but the Higgs data is again less spread and more sharply peaked. The range of Higgs data here is less than the QCD. This does not provide a ton of discrimination power, as the shapes and ranges are close.

phi - Phi seems randomly distributed, or if there is a pattern I need to look deeper. So as of now I do not see it as useful for discrimination between signal and backgorund.

mass - A useful amount of discrimination power here. Mass in the QCD data is distributed from 0 to 800, while the Higgs data is only from 20 to 160. They have peaks around the same alue, but for whatever reason the Higgs data lacks the long tail of the expected background. This might be used to minimize our background.



```python
def correlation_plot(data,feature,bins=40):
    
    plt.rcParams["figure.figsize"] = (50,12)
    
    fig, ax = plt.subplots(2, 7)
    
    fig.suptitle(f'Correlation Plots for {feature}', fontsize=16)
    
    for a in range(7):
        for i in range(2):
            
            ax[i][a].hist2d(data[feature],data[higgs.keys()[a + 7*i]],bins=bins)
            ax[i][a].set_xlabel(feature)
            ax[i][a].set_ylabel(str(higgs.keys()[a + 7*i]))
```

```python
for i in qcd.keys():
    correlation_plot(qcd,i)
```
